```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv(r"D:\PythonProjects\Data_Jobs\jobs_in_data.csv")
df.head()
```

```python
df.describe()
```

## Salary Distribution

```python
plt.figsize=(10, 6)
sns.histplot(df['salary_in_usd'], bins=20, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

## Job Titile and Salary

```python
# Job Title vs. Salary
plt.figure(figsize=(20, 10))
sns.barplot(x='job_title', y='salary_in_usd', data=df, ci=None)
plt.title('Average Salary by Job Title')
plt.xticks(rotation=90, ha='right', fontsize=11)
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.show()
```

## Job Category Comparassion

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='job_category', y='salary_in_usd', data=df, ci=None)
plt.title('Average Salary by Job Category')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Job Category')
plt.ylabel('Average Salary in USD')
plt.show()
```

## Experience Level and Salary

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

## Employment Type vs. Salary:

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='employment_type', y='salary_in_usd', data=df)
plt.title('Employment Type vs. Salary')
plt.xlabel('Employment Type')
plt.ylabel('Salary (USD)')
plt.show()
```

## Company Size Distribution:

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='company_size', data=df)
plt.title('Company Size Distribution')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.show()
```

## Experience Level vs. Salary:

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Experience Level vs. Salary')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

## Salary Currency Distribution:

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='salary_currency', data=df)
plt.title('Salary Currency Distribution')
plt.xlabel('Salary Currency')
plt.ylabel('Count')
plt.show()
```
